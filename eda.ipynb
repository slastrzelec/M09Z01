```python
from dotenv import load_dotenv
import os
import boto3
import pandas as pd
from io import StringIO

# 1Ô∏è‚É£ Wczytaj dane z .env
load_dotenv()

# 2Ô∏è‚É£ Utw√≥rz klienta boto3
session = boto3.session.Session()
client = session.client(
    's3',
    region_name=os.getenv('SPACES_REGION'),
    endpoint_url=f"https://{os.getenv('SPACES_REGION')}.digitaloceanspaces.com",
    aws_access_key_id=os.getenv('SPACES_KEY'),
    aws_secret_access_key=os.getenv('SPACES_SECRET')
)

# 3Ô∏è‚É£ Funkcja pomocnicza
def load_csv_from_space(bucket_name, file_key):
    obj = client.get_object(Bucket=bucket_name, Key=file_key)
    csv_data = obj['Body'].read().decode('utf-8')
    return pd.read_csv(StringIO(csv_data), sep=';')  # U≈ºyj ≈õrednika jako separatora

# 4Ô∏è‚É£ Wczytaj dane
df_2023 = load_csv_from_space('maratonwr', 'data/raw/halfmarathon_wroclaw_2023__final.csv')
df_2024 = load_csv_from_space('maratonwr', 'data/raw/halfmarathon_wroclaw_2024__final.csv')

# 5Ô∏è‚É£ PodglƒÖd
print("‚úÖ Dane 2023:")
display(df_2023.head(3))

print("‚úÖ Dane 2024:")
display(df_2024.head(3))
```

```python
# üîç Sprawd≈∫ liczbƒô i nazwy kolumn w obu plikach
print("üìä Liczba kolumn:")
print(f"2023: {len(df_2023.columns)}")
print(f"2024: {len(df_2024.columns)}\n")

# üîé Por√≥wnaj nazwy kolumn
cols_2023 = set(df_2023.columns)
cols_2024 = set(df_2024.columns)

# R√≥≈ºnice
only_in_2023 = cols_2023 - cols_2024
only_in_2024 = cols_2024 - cols_2023

print("üü¢ Kolumny wsp√≥lne:", len(cols_2023 & cols_2024))
print("üî¥ Tylko w 2023:", only_in_2023 if only_in_2023 else "Brak r√≥≈ºnic")
print("üîµ Tylko w 2024:", only_in_2024 if only_in_2024 else "Brak r√≥≈ºnic")
```

```python
print("üìã Czy kolejno≈õƒá kolumn jest identyczna?")
print(list(df_2023.columns) == list(df_2024.columns))
```

```python
# ‚ûï Dodaj kolumnƒô z rokiem do ka≈ºdego DataFrame
df_2023["Rok"] = 2023
df_2024["Rok"] = 2024

# üîó Po≈ÇƒÖcz dane w jeden zbi√≥r
df_all = pd.concat([df_2023, df_2024], ignore_index=True)

# üßπ (opcjonalnie) usu≈Ñ ewentualne duplikaty
df_all.drop_duplicates(inplace=True)

# üìä Sprawd≈∫ wynik
print("‚úÖ Po≈ÇƒÖczono dane:")
print(f"Liczba rekord√≥w: {len(df_all)}")
print(f"Liczba kolumn: {len(df_all.columns)}")
display(df_all.head(5))
```

```python
import pandas as pd

# üïí Funkcja konwertujƒÖca czas w formacie HH:MM:SS lub MM:SS na sekundy
def czas_na_sekundy(czas):
    try:
        if pd.isna(czas):
            return None
        czas = str(czas)
        parts = czas.split(":")
        if len(parts) == 3:
            h, m, s = map(int, parts)
            return h * 3600 + m * 60 + s
        elif len(parts) == 2:
            m, s = map(int, parts)
            return m * 60 + s
        else:
            return None
    except:
        return None

# üîé Wybieramy kolumny z nazwƒÖ zawierajƒÖcƒÖ "Czas"
czas_cols = [col for col in df_ml.columns if "Czas" in col]

# üîÑ Konwersja na sekundy
for col in czas_cols:
    df_ml[col + " [s]"] = df_ml[col].apply(czas_na_sekundy)

# üßπ Usuwamy stare kolumny tekstowe (czas w formacie 00:MM:SS)
df_ml.drop(columns=czas_cols, inplace=True)

# ‚úÖ PodglƒÖd zmian
print("üìä Typy danych po konwersji:\n", df_ml.dtypes)
print("\nüß≠ PodglƒÖd danych:")
print(df_ml.head(3))
```

```python
# üßπ Usu≈Ñ podw√≥jne oznaczenia jednostek i zbƒôdne spacje
df_ml.columns = df_ml.columns.str.replace(r'\s*\[s\]\s*\[s\]\s*', ' [s]', regex=True)
df_ml.columns = df_ml.columns.str.strip()

print("‚úÖ Nazwy kolumn po oczyszczeniu:")
print(df_ml.columns.tolist())
```
